```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
from functools import partial
%matplotlib inline
```

```python
def corr_month(month,HPD,df):
    MONTH = month
    df_m = df.copy()
    df_m = df_m[df_m.month==MONTH]*1

    corr_matrix = df_m.corr('spearman')#df.corr('kendall'),df.corr('kendall'),df.corr('spearman')
    if HPD not in corr_matrix:
        raise Exception("Input HPD must be included in HPD list.")
    hpd_cor = corr_matrix[HPD]#.sort_values()
    species = hpd_cor.loc[['Pelecaniformes', 'Gruiformes','Passeriformes',
                           'Suliformes', 'Ciconiiformes', 'Falconiformes',
                           'Charadriiformes', 'Anseriformes', 'Accipitriformes',
                           'Procellariiformes']]
    species.name = MONTH
    return species
```

```python
def get_corr(file):
    #file =r"/home/wb/Documents/virus_analysis/virus_dispersal/2.3.2.1_HPD_files/time_polygon_random/dataset_2.3.2.1_.csv"
    df = pd.read_csv(file).infer_objects()
    df.drop(['Unnamed: 0', 'point', 'lon_indx', 'lat_indx','lon', 'lat'],axis=1,inplace=True)
    df1 = df.iloc[:,1:19]*1
    for hpd in range(10,100,5):
        df1['HPD%s'%hpd] = df1[str(hpd/100)]*(hpd/100)
        #df1['HPD%s'%hpd] = df1['%s'%(hpd/100)]*(hpd/100)
        df1.drop([str(hpd/100)],axis=1,inplace=True)
    # got a mixed HPD value index
    df1['HPD'] = df1.iloc[:,0:19].apply(lambda x:
                           min(list(filter(lambda s:s!=0,x.to_list())))
                           if list(filter(lambda s:s!=0,x.to_list()))!=[]
                           else 0,axis=1)
    co_df = df.copy()
    co_df['HPD'] = df1['HPD']
    
    ## plot
    co_df.HPD.hist(bins=8)
    map_cor = partial(corr_month, HPD='HPD',df=co_df)
    month_stack = map(map_cor, list(range(1,13)))
    q = list(month_stack)
    stack_df = pd.concat(q,axis=1)
    return stack_df
```

```python
#file =r"/home/wb/Documents/virus_analysis/virus_dispersal/2.3.2.1_HPD_files/time_polygon_random/dataset_2.3.2.1_.csv"
file =r"/home/wb/Documents/virus_analysis/virus_dispersal/2.3.4.4_HPD_files/time_polygon_random/dataset_2.3.4.4_.csv"
stack_df = get_corr(file)
stack_df.T.plot(xticks = list(range(1,13)),figsize=(12,5))
plt.legend(loc=(1,0))
```

#### different season

```python
def cor_season(season,HPD,file):
    #HPD = 'HPD'
    #df = co_df
    df = pd.read_csv(file).infer_objects()
    df.drop(['Unnamed: 0', 'point', 'lon_indx', 'lat_indx','lon', 'lat'],axis=1,inplace=True)
    df1 = df.iloc[:,1:19]*1
    for hpd in range(10,100,5):
        df1['HPD%s'%hpd] = df1[str(hpd/100)]*(hpd/100)
        #df1['HPD%s'%hpd] = df1['%s'%(hpd/100)]*(hpd/100)
        df1.drop([str(hpd/100)],axis=1,inplace=True)
    #计算综合的HPD
    df1['HPD'] = df1.iloc[:,0:19].apply(lambda x:
                           min(list(filter(lambda s:s!=0,x.to_list())))
                           if list(filter(lambda s:s!=0,x.to_list()))!=[]
                           else 0,axis=1)
    co_df = df.copy()
    co_df['HPD'] = df1['HPD']
    df_m = co_df.copy()
    df_m = df_m[df_m.month.isin(season)]*1
    corr_matrix = df_m.corr('spearman')#df.corr('kendall'),df.corr('kendall'),df.corr('spearman')
    if HPD not in corr_matrix:
        raise Exception("Input HPD must be included in HPD list.")
    hpd_cor = corr_matrix[HPD]#.sort_values()
    species = hpd_cor.loc[['Pelecaniformes', 'Gruiformes','Passeriformes',
                               'Suliformes', 'Ciconiiformes', 'Falconiformes',
                               'Charadriiformes', 'Anseriformes', 'Accipitriformes',
                               'Procellariiformes']]
    #species.name = MONTH
    return species#.sort_values()
```

```python
#seasom
#file = r'..\data\dataset_2.3.4.4_.csv'
file = r'..\data\dataset_2.3.2.1_.csv'
Wintering = [12,1,2]
Fall_migration = [9,10,11]
Breeding = [5,6,7,8]
Spring_migration = [3,4]
season_l = [Spring_migration,Breeding,Fall_migration,Wintering]
out_list = []
for s in season_l:
    out_list.append(cor_season(s,"HPD",file))
temp = pd.concat(out_list,axis=1)
temp.columns=['Spring_migration','Breeding','Fall_migration','Wintering']
temp.T.plot(figsize=(12,5))
plt.legend(loc=(1,0))
```
