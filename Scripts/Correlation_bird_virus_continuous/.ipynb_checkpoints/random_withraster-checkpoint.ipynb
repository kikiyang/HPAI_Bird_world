```python
import os
import string
import csv
import re
import numpy as np
import pandas as pd
from osgeo import gdal

from time import time
from math import floor
from pandas import Timestamp
from dateutil.parser import parse#从字符串转为日期的格式
```

```python
orders = ['Pelecaniformes', 'Gruiformes', 'Passeriformes', 'Suliformes', 'Ciconiiformes',
 'Falconiformes', 'Charadriiformes', 'Anseriformes', 'Accipitriformes', 'Procellariiformes']
```

```python
def y3(x):
    y = []
    for i in range(1,13):
        y.append(x+str(i))
    return y

add_columns = list(np.array(list(map(y3, orders))).reshape([1,-1])[0])
```

```python
csvFile = open("./2.3.2.1Location.csv",'w',newline='',encoding='utf-8')
writer = csv.writer(csvFile)
csvRow = []

f = open("2.3.2.1Location.txt",'r',encoding='GB2312')
for line in f:
    csvRow = line.split()
    writer.writerow(csvRow)

f.close()
csvFile.close()
```

```python
random_df = pd.read_csv(r'2.3.2.1Location.csv')
random_df['lon']=random_df['long']
random_df.drop(['long'],inplace=True,axis=1)
```

```python
tif_path = r'D:\world virus&bird\map_raster/'
i = 'Accipitriformes1.tif'
env_data = gdal.Open(tif_path+i)#tiff数据
inf = env_data.GetGeoTransform()#tif坐标信息数据
raster_X = env_data.RasterXSize
raster_Y = env_data.RasterYSize
del env_data
```

```python
## get index
t1 = time()
   
random_df['lat']=pd.to_numeric(random_df['lat'])
random_df['lon']=pd.to_numeric(random_df['lon'])
          
random_df['lon_indx'] = random_df.lon.apply(lambda x:round((x-inf[0])/inf[1]))
random_df['lat_indx'] = random_df.lat.apply(lambda x:round((x-inf[3])/inf[5]))
             
random_df['lon_indx'] = random_df.lon_indx.apply(lambda x: raster_X-1  if x>=raster_X else x)
random_df['lat_indx'] = random_df.lat_indx.apply(lambda x: raster_Y-1  if x>=raster_Y else x)

print(time()-t1)
```

```python
raster_df = random_df.copy()
```

```python
'''
def sortkey(s):
    return re.findall(r"[-+]?\d*\.\d+|\d+",s)'''
```

```python
for file in add_columns:
    filename = os.path.splitext(file)[0]
    env_data = gdal.Open(os.path.join(tif_path,'%s.tif'%file))#tiff数据
    band = env_data.GetRasterBand(1).ReadAsArray()
    raster_df = raster_df.copy()
    raster_df[filename] = raster_df.apply(lambda x:band[x['lat_indx']][x['lon_indx']],axis=1)
```
