```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clade21 = pd.read_csv('WAI_data/2.3.2.1fAreaDTA_MJs_bi.txt')
clade34 = pd.read_csv('WAI_data/2.3.4.4fAreaDTA_MJs_bi.txt')
```

```python
clade21['dateinyear'] = clade21['date'] - np.floor(clade21['date'])
clade34['dateinyear'] = clade34['date'] - np.floor(clade34['date'])
```

```python
clade21['od'] = clade21['origin'] + '-' + clade21['destination']
clade34['od'] = clade34['origin'] + '-' + clade34['destination']
```

### Disribution of earliest virus lineage migration event in a year

```python
clade21_mindatInYr = clade21.groupby(['treeNumber','od'])['dateinyear'].min().reset_index()
clade21_mindatInYr['days']=clade21_mindatInYr['dateinyear']*365
clade21_mindatInYr = clade21_mindatInYr.sort_values('od')
```

```python
clade21_mindatInYr.groupby(['od'])['days'].describe(percentiles=[.025, .25, .5, .75, .975])
```

```python
fig21 = plt.figure(figsize=(30, 6))
plt.xticks(rotation=90)
bwp_21 = sns.boxplot(
    x="od", y="days", 
    data=clade21_mindatInYr, whis=[2.5,97.5])
plt.xlabel('Origin - Destination of virus lineage migration event',fontsize=12)
plt.ylabel('Earliest migration date (days in a year)',fontsize=12)
plt.tight_layout()
fig21.savefig('output/clade2.3.2.1_earliest_jumps.jpg')
```

```python
clade34_mindatInYr = clade34.groupby(['treeNumber','od'])['dateinyear'].min().reset_index()
clade34_mindatInYr['days']=clade34_mindatInYr['dateinyear']*365
clade34_mindatInYr = clade34_mindatInYr.sort_values('od')
```

```python
clade34_mindatInYr.groupby(['od'])['days'].describe(percentiles=[.025, .25, .5, .75, .975])
```

```python
fig34 = plt.figure(figsize=(30, 6))
plt.xticks(rotation=90)
bwp_34 = sns.boxplot(
    x="od", y="days", 
    data=clade34_mindatInYr, whis=[2.5,97.5])
plt.xlabel('Origin - Destination of virus lineage migration event',fontsize=12)
plt.ylabel('Earliest migration date (days in a year)',fontsize=12)
plt.tight_layout()
fig34.savefig('output/clade2.3.4.4_earliest_jumps.jpg')
```

### Virus lineage migration distributover months

```python
from scipy import stats
```

```python
fig, axes = plt.subplots(11,10,figsize=(80,80),dpi=100,sharex=True, sharey=True)

for i, (ax, od) in enumerate(zip(axes.flatten(), clade21.od.unique())):
    x = clade21.loc[clade21.od==od, 'dateinyear']
    
    res = stats.relfreq(x, numbins=12)
    x_space = res.lowerlimit + np.linspace(0, res.binsize*res.frequency.size,
                                 res.frequency.size)
    ax.bar(x_space,res.frequency,width=res.binsize)
    #ax.hist(x, alpha=0.5, bins=365, density=True, stacked=True, label=od)
    ax.set_title(od)
    ax.set_xlim([x_space.min(), x_space.max()])
    
plt.suptitle('Probability Histogram of date of Markov jump events', y=1.05, size=16)
#ax.set_xlim(0, 1); ax.set_ylim(0, 10);
#plt.tight_layout();
fig.savefig('WAI_data/MJ_Dis_Clade2.3.2.1_monthly.pdf')   # save the figure to file
plt.close(fig)    # close the figure window
```

```python
freq_clade21 = pd.DataFrame(columns=['origin_location','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
for i, (origin) in enumerate(clade21.origin.unique()):
    x=clade21.loc[clade21.origin==origin,'dateinyear']
    res=stats.relfreq(x,numbins=12)
    freql = res.frequency.tolist()
    freq_clade21.loc[i] = [origin] + freql

freq_clade21
```

```python
freq_clade21_dest = pd.DataFrame(columns=['dest_location','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
for i, (destination) in enumerate(clade21.destination.unique()):
    x=clade21.loc[clade21.destination==destination,'dateinyear']
    res=stats.relfreq(x,numbins=12)
    freql = res.frequency.tolist()
    freq_clade21_dest.loc[i] = [destination] + freql
    
freq_clade21_dest
```

```python
freq_clade34_dest = pd.DataFrame(columns=['dest_location','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
for i, (destination) in enumerate(clade34.destination.unique()):
    x=clade34.loc[clade34.destination==destination,'dateinyear']
    res=stats.relfreq(x,numbins=12)
    freql = res.frequency.tolist()
    freq_clade34_dest.loc[i] = [destination] + freql

freq_clade34_dest
```

```python
freq_clade34 = pd.DataFrame(columns=['origin_location','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
for i, (origin) in enumerate(clade34.origin.unique()):
    x=clade34.loc[clade34.origin==origin,'dateinyear']
    res=stats.relfreq(x,numbins=12)
    freql = res.frequency.tolist()
    freq_clade34.loc[i] = [origin] + freql

freq_clade34
```

```python
birdProb21 = pd.read_csv('WAI_data/2.3.2.1_bird.csv')
birdProb21.head()
```

```python
birdProb34 = pd.read_csv('WAI_data/2.3.4.4_bird.csv')
birdProb34.head()
```

```python
from IPython.core.debugger import set_trace
```

```python
corr_mj_bird_21_orig = pd.DataFrame(columns=['origin_location','bird_order','correlation_coefficient','p_value'])
i=0
for location in birdProb21.Location.unique():
    mj = freq_clade21.loc[freq_clade21.origin_location==location,['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]
    for bird in ['Accipitriformes','Anseriformes','Charadriiformes','Ciconiiformes','Falconiformes',
                 'Gruiformes','Passeriformes','Pelecaniformes','Procellariiformes','Suliformes']:
        birdProb = birdProb21.loc[birdProb21.Location==location,[bird+'1',bird+'2',bird+'3',bird+'4',bird+'5',bird+'6',bird+'7',bird+'8',bird+'9',bird+'10',bird+'11',bird+'12']]
        #cor = np.corrcoef(mj,birdProb)
        #set_trace()
        cor,p = stats.pearsonr(mj.values[0],birdProb.values[0])
        corr_mj_bird_21_orig.loc[i]= [location] + [bird] + [cor] + [p]
        #corr_mj_bird_21_orig.loc[i]= [location] + [bird] + [cor[0,1]] + [cor[0,0]]
        i=i+1
        
corr_mj_bird_21_orig
```

```python
corr_mj_bird_21_dest = pd.DataFrame(columns=['dest_location','bird_order','correlation_coefficient','p_value'])
i=0
for location in birdProb21.Location.unique():
    mj = freq_clade21_dest.loc[freq_clade21_dest.dest_location==location,['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]
    for bird in ['Accipitriformes','Anseriformes','Charadriiformes','Ciconiiformes','Falconiformes',
                 'Gruiformes','Passeriformes','Pelecaniformes','Procellariiformes','Suliformes']:
        birdProb = birdProb21.loc[birdProb21.Location==location,[bird+'1',bird+'2',bird+'3',bird+'4',bird+'5',bird+'6',bird+'7',bird+'8',bird+'9',bird+'10',bird+'11',bird+'12']]
        cor,p = stats.pearsonr(mj.values[0],birdProb.values[0])
        corr_mj_bird_21_dest.loc[i]= [location] + [bird] + [cor] + [p]
        i=i+1
        #set_trace()
        
corr_mj_bird_21_dest
```
